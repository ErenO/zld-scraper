```python
url_to_scrap = ['https://www.zalando.fr/manteaux-homme/?p=',
               'https://www.zalando.fr/vestes-homme/?p=',
               'https://www.zalando.fr/pulls-gilets-homme/?p=',
               'https://www.zalando.fr/sweatshirts-homme/?p=',
               'https://www.zalando.fr/t-shirts-polos-homme/?p=',
               'https://www.zalando.fr/t-shirts-tops-femme/?p=',
               'https://www.zalando.fr/sweatshirts-femme/?p=',
               'https://www.zalando.fr/pulls-gilets-femme/?p=',
               'https://www.zalando.fr/vestes-femme/?p=',
               'https://www.zalando.fr/manteaux-femme/?p=']
```

```python
imgs = {}
imgs['manteaux'] = []
imgs['veste'] = []
imgs['pull'] = []
imgs['sweatshirts'] = []
imgs['t-shirts'] = []
names = ['manteaux', 'veste', 'pull', 'sweatshirts', 't-shirts']
imgs_scrap = []
```

```python
for item in url_to_scrap:
    cls = ''
    print (item)
    for name in names:
        if (name in item):
            cls = name
            break;
    for i in range(1, 30):
        url = item + str(i)
        r = requests.get(url)
        soup = Bsoup(r.content)
        img_tags = soup.find_all('img')
        urls = [img['src'] for img in img_tags if 'jpg' in img['src']]
        imgs[cls].extend(urls)
    
```
